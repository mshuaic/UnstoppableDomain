```python
vaild_keys = set(filter(lambda x:x.split(".")[0]!="crypto", keys["keys"].keys()))
```

```python
vaild_keys
```

```python
def remove_keys(jsonstring, remaining_keys):
    data = eval(jsonstring)
    keys = set(remaining_keys) & set(data.keys())
    return {k:data[k] for k in keys}
    
    
```

```python
tmp['records_wo_cypto'] = tmp["records"].apply(remove_keys, remaining_keys=vaild_keys)
```
