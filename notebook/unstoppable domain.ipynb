```python
df["records"] = list(zip(df["from"], df["to"], df["blockNumber"]))
```

```python
df[["tokenId", "records"]]
```

```python
tokens = df[["tokenId", "records"]].groupby("tokenId")["records"].apply(list).reset_index(name="records")
tokens
```

```python
tokens["nTransfer"] = tokens["records"].apply(len)
```

```python
tokens[tokens["nTransfer"] > 1].sort_values("nTransfer")
```
