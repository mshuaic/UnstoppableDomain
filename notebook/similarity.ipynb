```python
import pandas as pd
```

```python
    topics = ["automotive", "travel", "science", "technology",
              "nature", "sports", "business", "beauty", "fashion", "music", "art",
              "movie", "entertainment", "media", "children", "health", "gambling",
              "religion", "education", "politics", "history", "war", "news",
              "food", "literature", "crime"]
    df = pd.read_csv("../data/similarity.csv", names=["domain"]+topics+["isTemplate"])
```

```python
df
```

```python
nonzero = df[(df[topics] != 0).any(axis=1)].reset_index(drop=True)
```

```python
result = nonzero[nonzero["isTemplate"] == False].drop_duplicates()
```

```python
hashoshi = result[result["domain"] == "hashoshi.crypto"][topics].iloc[0].sort_values().plot.barh(figsize=(7, 7))
```
